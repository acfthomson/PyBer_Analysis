```python
# Import dependencies
import matplotlib.pyplot as plt
import statistics
```

```python
# Ridesharing data
# Set the x-axis to a list of strings for each month
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", 
          "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# MATLAB method
# Create the plot and add a label for the legend
plt.plot(x_axis, y_axis, label='Boston')

# Create labels for the x and y axes
plt.xlabel("Date")
plt.ylabel("Fare($)")

# Set the y limit between 0 and 45
plt.ylim(0, 45)

# Create a title
plt.title("PyBer Fare by Month")

# Add a grid
plt.grid()

# Add a legend
plt.legend()
```

```python
# Get the standard deviation of the values in the y-axis
stdev = statistics.stdev(y_axis)
stdev
```

```python
# Add the errorbar() function, which will show either standard deviation, standard error, 
# confidence intervals, or minimum and macimum values of a dataset
plt.errorbar(x_axis, y_axis, yerr=stdev, capsize=3)
```

```python
# Object-oriented method
# Create a plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()

# Create a plot with a legend label called 'Boston'.  Legend line color is green 
# with a diamond marker
ax.plot(x_axis, y_axis, label='Boston', color='green', marker='d')

# Create labels for the x- and y-axes
ax.set_xlabel("Date")
ax.set_ylabel("Fares($)")

# Set the y limit between 0 and 45
ax.set_ylim(0, 45)

# Create a title
ax.set_title("PyBer Fare by Month")

# Add a grid
ax.grid()

# Add a legend
ax.legend()

# Change line width to 2 pixels
ax.plot(x_axis, y_axis, linewidth=2, marker='d')

# Change color of plot line to green
ax.properties()['children'][1].set_color('green')
```

```python
# Add error bars using the object-oriented approach
fig, ax = plt.subplots()
ax.errorbar(x_axis, y_axis, yerr=stdev, capsize=3)
plt.show()
```

```python
# Create a bar chart using the MATLAB approach
plt.bar(x_axis, y_axis, color='green', label='Boston')

# Create labels for the axes
plt.xlabel("Date")
plt.ylabel("Fare($)")

# Create a title
plt.title('PyBer Fare by Month')

# Add the legend
plt.legend()
```

```python
# Add error bars to the bar chart
plt.bar(x_axis, y_axis, yerr=stdev, capsize=3)
```

```python
# Create a bar chart with error bars using the object-oriented method
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis, yerr=stdev, capsize=3)
```

```python
# Create a horizontal bar chart with the MATLAB method and major ticks at every $5 increment
# To adjust ajor x-axis ticks on a horizontal bar chart, use xticks()
import numpy as np
plt.barh(x_axis, y_axis)
plt.xticks(np.arange(0, 51, step=5.0))
plt.gca().invert_yaxis()
```

```python
# Create a horizontal bar chart with the object-oriented method and major ticks at every $5 increment
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis)
ax.set_xticks(np.arange(0, 51, step=5.0))
plt.show()
```

```python
# Add minor ticks using the object-oriented method
from matplotlib.ticker import MultipleLocator
# Increase the size of the plot figure
fig, ax = plt.subplots(figsize=(8,8))
ax.barh(x_axis, y_axis)
ax.set_xticks(np.arange(0, 51, step=5.0))

# Create minor ticks at an increment of 1.
ax.xaxis.set_minor_locator(MultipleLocator(1))
plt.show()
```
