```python
# Import dependencies
import matplotlib.pyplot as plt
```

```python
# Ridesharing data
# Set the x-axis to a list of strings for each month
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
# Create a figure
fig, ax = plt.subplots()

# Define the axes
ax.plot(x_axis, y_axis)
```

```python
# Create a plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
plt.show()
```

```python
# MATLAB method

# Create the plot and add a label for the legend
plt.plot(x_axis, y_axis, label='Boston')

# Create labels for the x and y axes
plt.xlabel("Date")
plt.ylabel("Fare($)")

# Set the y limit between 0 and 45
plt.ylim(0, 45)

# Create a title
plt.title("PyBer Fare by Month")

# Add a grid
plt.grid()

# Add a legend
plt.legend()
```

```python
# Object-oriented method
# Create a plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()

# Create a plot with a legend label called 'Boston'.  Legend line color is green with a diamond marker
ax.plot(x_axis, y_axis, label='Boston', color='green', marker='d')

# Create labels for the x- and y-axes
ax.set_xlabel("Date")
ax.set_ylabel("Fares($)")

# Set the y limit between 0 and 45
ax.set_ylim(0, 45)

# Create a title
ax.set_title("PyBer Fare by Month")

# Add a grid
ax.grid()

# Add a legend
ax.legend()

# Change line width to 2 pixels
ax.plot(x_axis, y_axis, linewidth=2, marker='d')

# Change color of plot line to green
ax.properties()['children'][1].set_color('green')
```

```python
# Create a bar chart
plt.bar(x_axis, y_axis, color='green', label='Boston')

# Create labels for the axes
plt.xlabel("Date")
plt.ylabel("Fare($)")

# Create a title
plt.title('PyBer Fare by Month')

# Add the legend
plt.legend()
```

```python
# Create a horizontal bar chart using the MATLAB method
plt.barh(x_axis, y_axis, color='magenta', label='Boston')

# Put months in top-down order
# gca() method --> get current axes
plt.gca().invert_yaxis()

# Add a legend
plt.legend()

# Add titles to x- and y-axes
plt.xlabel("Fare($)")
plt.ylabel("Date")

# Add a graph title
plt.title("PyBer Fare by Month")
```

```python
# Create a bar chart using the object-oriented method
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis)
```

```python
# Create a horizontal bar chart using the object-oriented method
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis, color='cyan', label='Chicago')

ax.set_ylabel("Date")
ax.set_xlabel("Fares($)")

# Add a legend
ax.legend()

# Create a title
ax.set_title("PyBer Fare by Month")
```

```python
# Create a scatter plot using the MATLAB method
plt.scatter(y_axis, x_axis, c='red', label='Chicago')

# Set limit for x-axis
plt.xlim(0, 50)

# Add a legend
plt.legend()

# Add titles to x- and y-axes
plt.xlabel("Fare($)")
plt.ylabel("Date")

# Add a graph title
plt.title("PyBer Fare by Month")
```

```python
# Create a bubble chart using MATLAB method
# Create the plot where the marker is equal to the data float value of the y-axis
#plt.scatter(x_axis, y_axis, s=y_axis)

# Make data points larger, but relative to data float value of y-axis
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)
    
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
# Create same bubble chart as above cell but use list comprehension
plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```

```python
# Create scatter plot using object-oriented method
fix, ax = plt.subplots()
ax.scatter(x_axis, y_axis)
```

```python
# Create a bubble chart using object-oriented method
fig, ax = plt.subplots()

# Change markers to 5 times each data point
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*5)

# Create the plot
ax.scatter(y_axis, x_axis, s=y_axis_larger, c='blue', alpha=0.5, label='Boston', 
           edgecolors='black')

# Ser limit on x-axis
ax.set_xlim(0, 50)

# Create labels for the x- and y-axes
ax.set_xlabel("Fares($)")
ax.set_ylabel("Date")

# Create a title
ax.set_title("PyBer Fare by Month")

# Add a legend
ax.legend()
```

```python
# Create a pie chart using MATLAB method

# Assign 12 colors, one for each month
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow",
          "orange", "gold", "indianred", "tomato", "mistyrose"]

# Explode values using array-like data points
# Length of typle should be equal to number of wedges
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)

# Increase size of pie chart
plt.subplots(figsize=(8,8))

# autopct='%.1f%%' --> add percentage to pie chart with one decimal place
plt.pie(y_axis, 
        explode=explode_values, 
        colors=colors,
        labels=x_axis, 
        autopct='%.1f%%')

plt.show()
```

```python
# Create a pie chart using the object-oriented method

# Assign 12 colors, one for each month
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow",
          "orange", "gold", "indianred", "tomato", "mistyrose"]

# Explode values using array-like data points
# Length of typle should be equal to number of wedges
explode_values = (0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)

# Increase size of pie chart
fig, ax =plt.subplots(figsize=(8,8))

ax.pie(y_axis,
       explode=explode_values,
       colors=colors,
       labels=x_axis,
       autopct='%.1f%%')

plt.show()
```
